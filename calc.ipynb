```
import numpy as np
import matplotlib.pyplot as plt
import math
import numpy.linalg as LA
```

```
H = np.array([[0, -math.sqrt(3)],[-math.sqrt(3), 0] ])
H
```

```
[v,d] = LA.eig(H)
print(v)
print(d)
```

```
H = np.array([[0, -math.sqrt(3)],[-math.sqrt(3), 2] ])
H
```
