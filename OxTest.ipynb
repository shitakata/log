温度で

```
lr = LinearRegression()

X = dataset['Temp'].values    # 説明変数
Y = dataset['Ox Th'].values    # 目的変数

lr.fit(X.reshape(-1, 1), Y.reshape(-1, 1))         
```

```
print(lr.coef_, lr.intercept_)
```

```
print(lr.score(X.reshape(-1, 1), Y.reshape(-1, 1)))
```

```
scross = []

score = cross_val_score(lr, X, Y, scoring = 'r2', cv = 5).mean()
score
```

```
plt.scatter(X,Y)
plt.plot(X.reshape(-1,1), lr.predict(X.reshape(-1,1)), color = 'r')
```

重回帰

```
clf = linear_model.LinearRegression()
 
# 説明変数exval
exval = dataset.drop('Ox Th', axis=1)
X = exval
 
# 目的変数Ox Th
Y = dataset['Ox Th']
 
# 予測モデルを作成
clf.fit(X, Y)
 
# 偏回帰係数
print(pd.DataFrame({"Name":exval.columns,
                    "Coefficients":clf.coef_}).sort_values(by='Coefficients') )
 
# 切片 (誤差)
print(clf.intercept_)
```

```
print(clf.score(X, Y))
```

```
scross = []

score = cross_val_score(clf, X, Y, scoring = 'r2', cv = 5).mean()
score
```
